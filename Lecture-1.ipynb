## Python's variable (data) types
- Numberic types (integers, floats, complex numbers), 
- Boolean types (True or False),
- Sequence types (strings, lists, tuples and range),
- Mapping types (dictionaries)
- Set types (sets and frozen sets)

## Numeric types

### Example
- Integer: 42, -7, 0, 1000, 0xFF (hexadecimal representation)
- Float: 3.14, 1.0, 2.71828
- Complex: 1+2j, 3-4j, 1j
